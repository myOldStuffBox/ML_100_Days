```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 作業
1. 請將 app_train 中的 CNT_CHILDREN 依照下列規則分為四組，並將其結果在原本的 dataframe 命名為 CNT_CHILDREN_GROUP
    * 0 個小孩
    * 有 1 - 2 個小孩
    * 有 3 - 5 個小孩
    * 有超過 5 個小孩

2. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，列出各組的平均 AMT_INCOME_TOTAL，並繪製 boxplot
3. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，對 AMT_INCOME_TOTAL 計算 [Z 轉換](https://en.wikipedia.org/wiki/Standard_score) 後的分數

```python
#1
"""
Your code here
"""
cut_rule = [-100,0,2,5,100]      

app_train['CNT_CHILDREN_GROUP'] = pd.cut(app_train['CNT_CHILDREN'].values, cut_rule)
'''
pd.cut(x, bins)
x : 為要分組資料
bins : 為區間( 下區間 < X <= 上區間 )

此區間為 [
            -100 < X <= 0   (0)
            0 < X <= 2    (1-2)
            2 < X <= 5    (3-5)
            5 < X <= 100  ( 5<X )
        ]

'''
app_train['CNT_CHILDREN_GROUP'].value_counts()
#print(app_train['CNT_CHILDREN_GROUP'])
```

```python
#2-1    # 利用 ['CNT_CHILDREN_GROUP','TARGET'] ，對['AMT_INCOME_TOTAL']進行分類
"""
Your code here
"""
grp = ['CNT_CHILDREN_GROUP','TARGET']

grouped_df = app_train.groupby(grp)['AMT_INCOME_TOTAL']
grouped_df.mean()
```

```python
#2-2   產生['AMT_INCOME_TOTAL']被分類後的boxplot圖
"""
Your code here
"""
plt_column = 'AMT_INCOME_TOTAL'
plt_by = ['CNT_CHILDREN_GROUP','TARGET']

app_train.boxplot(column=plt_column, by = plt_by, showfliers = False,  figsize=(12,12) )
'''
df物件.boxplot(column, by )
column : 要使用哪欄數值做圖
by : 要用哪欄做分類(不同類別會有獨立的圖，欄中相同的資料視為一類)
showfliers : 似乎是[是否顯示離群直]
figsize : 產生圖片大小
'''

plt.suptitle('')
plt.show()
```

```python
#3   z-標準化被分類後的['AMT_INCOME_TOTAL']
"""
Your code here
"""
app_train['AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET'] = grouped_df.apply(lambda x: (x-x.mean()) / x.std() )

app_train[['AMT_INCOME_TOTAL','AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET']].head(8)
```

```python
# 用z-標準化後的資料，依照 ['CNT_CHILDREN_GROUP','TARGET'] 分類並繪圖

app_train.boxplot(column = 'AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET' ,
                  by = ['CNT_CHILDREN_GROUP','TARGET'],
                  showfliers = False,
                  figsize = (12,12) 
                 )
```

```python
# 將 grouped_df 進行 MinMaxScaler 作圖

app_train['MinMax'] = grouped_df.apply(lambda x : (x-x.min()) / (x.max() - x.min()) )
#print(app_train['MinMax'].head(8))

app_train.boxplot(column='MinMax',
                  by =  ['CNT_CHILDREN_GROUP','TARGET'] ,
                  showfliers = False,
                  figsize=(12,12)
                 )
```
