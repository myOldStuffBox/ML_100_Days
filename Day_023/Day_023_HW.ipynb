```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
import copy                     # 用於之後的 copy.deepcopy()
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

data_path = 'data/'
df_train = pd.read_csv(data_path + 'titanic_train.csv')
df_test = pd.read_csv(data_path + 'titanic_test.csv')

# 對資料做分配，合併 train、test 資料
train_Y = df_train['Survived']                      # 取出目標值
ids = df_test['PassengerId']                        # 取出乘客ID
df_train = df_train.drop(['PassengerId', 'Survived'] , axis=1)
df_test = df_test.drop(['PassengerId'] , axis=1)
df = pd.concat([df_train,df_test])                  # 合併 train、test 資料
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')

# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]           # 剩數值型欄位
df = df.fillna(0)               # NAN 值填滿 0
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]    # 得到 train 數據資料數
df.head()
```

```python
# 顯示 Fare 與目標值的散佈圖    // 未經處理
import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(df['Fare'][:train_num])
plt.show()
```

```python
# 計算基礎分數
df_mm = MMEncoder.fit_transform(df)                         # 最大最小化數據
train_X = df_mm[:train_num]                                 # 只取出 train 資料做訓練
estimator = LogisticRegression()                            # 回歸模型
cross_val_score(estimator, train_X, train_Y, cv=5).mean()   # 計算分數
```

# 作業1 
* 試著在鐵達尼的票價 (Fare) 欄位中使用對數去偏 (log1p) , 結果是否更好?

```python
# 將 Fare 取 log1p 後, 看散佈圖, 並計算分數
df_fixed = copy.deepcopy(df)                                # 深層複製 : copy.deepcopy() : 可完整複製出一個獨立新變數，如用copy() 只會有第一層位置不同 
"""
Your Code Here, transform df_fixed['Fare'] with log1p
"""
df_fixed['Fare'] = np.log1p(df_fixed['Fare'])               # 做 log1p(x)
sns.distplot(df_fixed['Fare'][:train_num])                  # [:train_num] 只印出 train 數據的圖形
plt.show()

df_fixed = MMEncoder.fit_transform(df_fixed)
train_X = df_fixed[:train_num]
estimator = LogisticRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()   # 計算新的分數
```

# 作業2
* 最後的 boxcox 區塊直接執行會造成錯誤, 起因為輸入值有負值, 請問如何修正後可以使用 boxcox? (Hint : 試圖修正資料)

```python
# 將 Fare 取 boxcox 後, 看散佈圖, 並計算分數 (執行會有 error, 請試圖修正)   // error 原因為有人 'Fare' == 0
from scipy import stats
df_fixed = copy.deepcopy(df)
"""
Your Code Here, fix the error
"""
df_fixed['Fare'] = df_fixed['Fare'].replace( {0 : 0.001} )      # 將 0 取代成 0.001
'''
其他人作法 : 
            有人全部加上 0.001 or 0.1
            有人用 np.where( condition, x, y) : 滿足 condition條件 輸出[x]，不滿足輸出[y]
            他的設置為 : ( df_fixed['Fare'] <= 0, 0.001, df_fixed['Fare'] )     // y 為 返回原值(df_fixed['Fare'])
'''

df_fixed['Fare'] = stats.boxcox(df_fixed['Fare'], lmbda=0.15)
sns.distplot(df_fixed['Fare'][:train_num])
plt.show()

df_fixed = MMEncoder.fit_transform(df_fixed)
train_X = df_fixed[:train_num]
estimator = LogisticRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```
